# 📅 Day 10 – Model Evaluation, Feature Importance & Hyperparameter Tuning

## 🚀 Overview

Day 10 focused on improving model evaluation, understanding feature importance, and introducing hyperparameter tuning techniques. The goal was to move beyond basic model building and start interpreting models in a business context.

---

## 🎯 Key Learnings

### 1. 📊 Feature Importance

* Used Random Forest to identify the most influential features affecting predictions.
* Helped bridge the gap between model output and business insights.

#### 🔍 Key Takeaways:

* Features like **tenure**, **MonthlyCharges**, and **Contract type** significantly impact churn.
* Feature importance helps prioritize business strategies.

---

### 2. 📈 ROC-AUC Score

* Learned to evaluate model performance beyond accuracy.

#### 🔍 Why ROC-AUC?

* Measures model’s ability to distinguish between classes.
* Works well even with imbalanced datasets.

#### 📌 Interpretation:

* 0.5 → Random model
* 0.7–0.8 → Good
* 0.8+ → Strong model

---

### 3. ⚙️ Hyperparameter Tuning with GridSearchCV

* Introduced automated tuning of model parameters.

#### 🔍 What it does:

* Tests multiple parameter combinations
* Selects the best performing model

#### 📌 Example Parameters:

* `n_estimators`
* `max_depth`

---

### 4. 🧠 Model Interpretation → Business Insights

Converted model outputs into actionable insights:

#### 💡 Example:

* **Insight:** Customers with low tenure are more likely to churn
* **Action:** Improve onboarding and early engagement

---

## 🔥 Key Skills Developed

* Model evaluation using **F1-score & ROC-AUC**
* Understanding **feature importance**
* Basic **hyperparameter tuning**
* Translating ML results into **business recommendations**

---

## 🧠 Important Concepts

* Accuracy alone is not sufficient for classification
* F1-score provides a balance between precision and recall
* Feature importance helps in understanding model behavior
* Hyperparameter tuning improves model performance

---

## 🏁 Conclusion

Day 10 marked a shift from simply building models to:

* Evaluating them effectively
* Understanding what drives predictions
* Connecting technical results to business impact
