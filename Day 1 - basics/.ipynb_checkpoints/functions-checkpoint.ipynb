#### Task 1: Add 2 numbers

```python
def addition(a,b):
    return a+b
```

#### Task 2: Multiply 2 numbers

```python
def multiply(a,b):
    return a*b
```

#### Task 3: Highest number from list

```python
def highest(numbers):
    max_num = numbers[0]
    for digit in numbers:
        if digit > max_num:
            max_num = digit
    return max_num
```

#### Task 4: Even number count from list

```python
def count_even_number(numbers):
    even_digit = 0
    for digit in numbers:
        if digit%2 == 0:
            even_digit += 1
    return even_digit
```

```python
print("Sum: ", addition(5,12))
print("Product: ", multiply(5,12))
numbers = [10,22,4,44,12,1,-11,-220]
print("Highest number: ", highest(numbers))
print("Even Count: ",count_even_number(numbers))
```
