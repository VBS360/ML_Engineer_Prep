#### Task 1: Find average of a list

```python
def calculate_average(numbers):
    if len(numbers) == 0:
        return 0
    total = 0
    for num in numbers:
        total += num
    average = total / len(numbers)
    return round(average, 2)
```

#### Task 2: Remove negative numbers from list

```python
def remove_negatives(numbers):
    clean_list = []
    for num in numbers:
        if num >= 0:
            clean_list.append(num)
    return clean_list
```

#### Task 3: Create squared list using list comprehension

```python
def squared_list(numbers):
    return [num**2 for num in numbers]
```

```python
numbers = [5,13,-22,18,31,-4,60]

print("Average of list: ",calculate_average(numbers))
print("Cleaned list: ",remove_negatives(numbers))
print("Squared list: ",squared_list(numbers))
```
