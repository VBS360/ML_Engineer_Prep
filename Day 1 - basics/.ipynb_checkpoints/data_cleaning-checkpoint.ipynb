#### Task 1: Create list with None values

```python
data = [10, None, 25, None, 30]
```

#### Task 2: Remove None values

```python
def cleaned_data(data):
    clean_data = []
    for value in data:
        if value is not None:
            clean_data.append(value)
    return clean_data
```

#### Task 3: Print cleaned data

```python
print("Cleaned data: ",cleaned_data(data))
```
