#### Task 1: Create student dataset (list of dictionaries)

```python
students = [
    {"name": "Abhay", "marks": 86},
    {"name": "Suraj", "marks": 56},
    {"name": "Mahima", "marks": 79},
    {"name": "Shivay", "marks": 80},
    {"name": "Falgun", "marks": 92},
    {"name": "Rashi", "marks": 91}
]
```

#### Task 2: Print all student names

```python
def get_student_names(students):
    names = []
    for student in students:
        names.append(student["name"])
    return names
```

#### Task 3: Find average marks

```python
def calculate_average_marks(students):
    if len(students) == 0:
        return 0
    
    total = 0
    for student in students:
        total += student["marks"]
    
    return total / len(students)
```

#### Task 4: Print students with marks > 80

```python
def students_above_80(students):
    result = []
    for student in students:
        if student["marks"] > 80:
            result.append(student["name"])
    return result
```

```python
print("Student Names:", get_student_names(students))
print("Average Marks:", calculate_average_marks(students))
print("Above 80:", students_above_80(students))
```
