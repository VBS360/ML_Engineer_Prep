# 📊 Statistics Round

## 🔹 Q1: What is the difference between bias and variance?

**Answer:**

Bias is error due to overly simple assumptions, leading to underfitting. Variance is error due to high sensitivity to training data, leading to overfitting. The goal is to balance both for good generalization.

---

## 🔹 Q2: When would you use median instead of mean?

**Answer:**

Median is used when the dataset has outliers or is skewed, as it is robust to extreme values. Mean is used when the data is normally distributed without significant outliers.

---

## 🔹 Q3: What is correlation and how is it useful?

**Answer:**

Correlation measures the strength and direction of the relationship between two variables. In machine learning, it is useful for feature selection, as highly correlated features can be redundant.

---

## 🔹 Q4: What is p-value?

**Answer:**

The p-value is the probability of observing the data assuming the null hypothesis is true. A low p-value indicates strong evidence against the null hypothesis, so we reject it.

# 🤖 Machine Learning Technical Round

## 🔹 Q1: Why is accuracy not a good metric for imbalanced datasets?

**Answer:**

Accuracy can be misleading in imbalanced datasets because the model may predict only the majority class and still achieve high accuracy.

In such cases, metrics like precision, recall, and F1-score are more appropriate, especially when detecting the minority class is important.

---

## 🔹 Q2: What would you do if your deployed model performance degrades?

**Answer:**

First, monitor performance metrics to confirm degradation. Then check for data drift or concept drift by comparing new data with training data.

If drift is present, retrain the model with updated data. Also validate input quality and adjust features or model parameters if needed.

---

## 🔹 Q3: How do you choose between Logistic Regression and Random Forest?

**Answer:**

Start with Logistic Regression as a baseline due to simplicity and interpretability. If the data has complex patterns, move to Random Forest.

Compare models using cross-validation. Choose Logistic Regression for interpretability if performance is similar, otherwise choose Random Forest for better accuracy.

---

## 🔹 Q4: How do you provide probability instead of prediction?

**Answer:**

Use the `predict_proba()` method instead of `predict()`.

Example:

```python
model.predict_proba(input_data)[0][1]
```

This returns the probability of the positive class.

---

## 🔹 Q5: How can you handle class imbalance?

**Answer:**

* Class weighting (class_weight='balanced')
* Resampling (SMOTE, undersampling)
* Threshold tuning
* Using appropriate metrics (F1-score, recall)
* Ensemble methods like balanced random forest
