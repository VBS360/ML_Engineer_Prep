```python
import pandas as pd
```

```python
store = pd.read_csv("store.csv")
train = pd.read_csv("train.csv")
store_sales = train.merge(store, on='Store', how='left')
store_sales.head()
```

```python
store_sales.describe()
```

```python
store_sales.info()
```

```python
store_sales.isnull().sum()
```

```python
missing = (store_sales.isnull().sum()/len(store_sales)) * 100
missing.sort_values(ascending=False)
```

```python
store_sales['PromoInterval'].fillna('None',inplace=True)
store_sales['Promo2SinceYear'].fillna(0,inplace=True)
store_sales['Promo2SinceWeek'].fillna(0,inplace=True)
store_sales['CompetitionOpenSinceYear'].fillna(0,inplace=True)
store_sales['CompetitionOpenSinceMonth'].fillna(0,inplace=True)
store_sales['CompetitionDistance'].fillna(0,inplace=True)
```
