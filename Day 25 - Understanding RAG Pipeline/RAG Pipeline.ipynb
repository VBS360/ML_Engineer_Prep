# -------------------------------
# SIMPLE RAG PIPELINE (TF-IDF BASED)
# -------------------------------

#### We are using TF-IDF to convert text into vectors (numerical representation)and cosine similarity to retrieve the most relevant document.

This project demonstrates a minimal Retrieval-Augmented Generation (RAG) pipeline using TF-IDF embeddings and cosine similarity.

## Features

* Converts documents into vector representations
* Retrieves most relevant document based on query
* Demonstrates basic retrieval logic used in RAG systems

## Tech Stack

* Python
* scikit-learn

## Use Case

This is a simplified version of how RAG systems retrieve context before generating responses in LLM-based applications.

# -------------------------------
# Step 1: Import Libraries
# -------------------------------

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# -------------------------------
# Step 2: Define Documents (Knowledge Base)
# -------------------------------

```python
documents = [
    "Diabetes is a chronic condition affecting blood sugar",
    "Exercise improves mental health",
    "Healthy diet is essential for good health",
    "Stress can impact physical well-being"
]
```

#### These are the documents that our system will search from.
#### In real RAG systems, this could be a database or vector store.

# -------------------------------
# Step 3: Convert Text → Vectors (Embedding)
# -------------------------------

#### TF-IDF (Term Frequency - Inverse Document Frequency):
#### - Converts each document into a numerical vector
#### - Words that are important in a document but rare overall get higher weight

```python
vectorizer = TfidfVectorizer()

# Fit + transform:
# - Learn vocabulary from documents
# - Convert each document into vector form

doc_vectors = vectorizer.fit_transform(documents)
```

#### Now:
#### Each document = a vector in high-dimensional space
#### Shape example: (num_documents, num_unique_words)

# -------------------------------
# Step 4: Define Query Function (Core RAG Logic)
# -------------------------------

```python
def rag_query(query):
    """
    This function simulates a RAG pipeline:
    1. Convert query to vector
    2. Compare with document vectors
    3. Retrieve most similar document
    """

    # Convert query into same vector space
    query_vector = vectorizer.transform([query])

    # Compute similarity between query and all documents
    # Cosine similarity measures angle between vectors:
    # closer to 1 = more similar
    similarities = cosine_similarity(query_vector, doc_vectors)

    # Get index of most relevant document
    best_match_index = similarities.argmax()

    # Return best matching document
    return documents[best_match_index]
```

# -------------------------------
# Step 5: Test the Pipeline
# -------------------------------

```python
query_1 = "How to control blood sugar?"
query_2 = "How to stay mentally healthy?"

print("Query:", query_1)
print("Best Match:", rag_query(query_1))

print("\nQuery:", query_2)
print("Best Match:", rag_query(query_2))
```
